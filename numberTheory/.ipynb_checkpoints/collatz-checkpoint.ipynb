```python
def collatz(num): #Returns the collatz number of num. i.e. then number of collatz iterations it takes for 
    count = 1
    while(num!=1):
        if num%2 == 1:
            num = num*3 + 1
            count+=1
        else:
            num = num/2
            count+=1
    return count
```

```python
import matplotlib.pyplot as plt

num = []
col = []
for i in range(1,5000):
    num.append(i)
    col.append(collatz(i))

fig = plt.figure()  
plt.figure(figsize=(15, 10))
plt.plot(num, col,'ro')
plt.xlabel('n')
plt.ylabel('collatz(n)')
plt.title("Collatz(n) vs. n")
plt.show()
```
