```python
def digital_root(n):
    #returns the digital root of n (recursive single digit sum of n's digits)
    while len(str(n))>1:
        s = 0
        for i in str(n): s += int(i)
        n = s
    return n

print(digital_root(65536)) #6+5+5+3+6 = 25, 2+5 = 7

def digital_root_mult(n):
    #returns the digital root of n (recursive single digit product of n's digits)
    while len(str(n))>1:
        s = 1
        for i in str(n): s *= int(i)
        n = s
    return n

print(digital_root_mult(4321)) #4*3*2*1 = 24, 2*4 = 8
```

```python
import matplotlib.pyplot as plt

num = []
dr = []
for i in range(1,101):
    num.append(i)
    dr.append(digital_root(i))

fig = plt.figure()  
plt.figure(figsize=(15, 10))
plt.plot(num, dr,'r')
plt.xlabel('n')
plt.ylabel('digital root(n)')
plt.title("Digital Root(n) vs. n")
plt.show()
```

```python
num = []
dr = []
for i in range(1,210):
    num.append(i)
    dr.append(digital_root_mult(i))

fig = plt.figure()  
plt.figure(figsize=(15, 10))
plt.plot(num, dr,'r')
plt.xlabel('n')
plt.ylabel('digital root mult(n)')
plt.title("Digital Root Mult(n) vs. n")
plt.show()
```
