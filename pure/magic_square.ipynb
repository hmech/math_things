# magic squares
1. best numbers to add to are 148-152 and all numbers in that visinity 
2. Thus use sets[142]-sets[146]

```python
sets = []
for i in range(1,100):
    for j in range(i+1,100):
        for k in range(j+1,100):
            found = True
            for m in range(len(sets)):
                if sets[m][0] == i+j+k:
                    sets[m].append([i,j,k])
                    found = False
            if found:
                sets.append([i+j+k,[i,j,k]])
```

```python
def trip_perm(trip):
    ret = []
    perms = [[0,1,2],[0,2,1],[1,0,2],[1,2,0],[2,0,1],[2,1,0]]
    for i in perms:
        temp = []
        for k in i:
            temp.append(trip[k])
        ret.append(temp)
    return ret
```

```python
print(sets[145][0])
for i in range(1,len(sets[145])):
    print(trip_perm(sets[145][i]))
```
